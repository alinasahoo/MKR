## Step 2 : Data Cleaning

```python
## Data Cleaning
import re
for column in hiv_data1.columns:
    hiv_data1['Number of New HIV Infections'] = hiv_data1['Number of New HIV Infections'].str.replace(r"\[.*\]", "")
    hiv_data1['New HIV Infections - Both Sexes'] = hiv_data1['New HIV Infections - Both Sexes'].str.replace(r"\[.*\]", "")
    hiv_data1['New HIV Infections - Male'] = hiv_data1['New HIV Infections - Male'].str.replace(r"\[.*\]", "")
    hiv_data1['New HIV Infections - Female'] = hiv_data1['New HIV Infections - Female'].str.replace(r"\[.*\]", "")
```

```python
# Cleaning special characters

spec_chars = ["&",";","l","t"]

for char in spec_chars:
    hiv_data1['Number of New HIV Infections'] = hiv_data1['Number of New HIV Infections'].str.replace(' ','')
    hiv_data1['New HIV Infections - Both Sexes'] = hiv_data1['New HIV Infections - Both Sexes'].str.replace(char, ' ')
    hiv_data1['New HIV Infections - Male'] = hiv_data1['New HIV Infections - Male'].str.replace(char,' ')
    hiv_data1['New HIV Infections - Female'] = hiv_data1['New HIV Infections - Female'].str.replace(char, ' ')
```
