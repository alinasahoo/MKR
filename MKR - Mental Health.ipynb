# Step 2 : Data Cleaning

```python
#dealing with missing data
#Let’s get rid of the variables "Timestamp",“comments”, “state” just to make our lives easier.
health = health.drop(['comments'], axis= 1)
health = health.drop(['state'], axis= 1)
health = health.drop(['Timestamp'], axis= 1)

health.isnull().sum().max() #just checking that there's no missing data missing as of now
health.head(5)
```

## Cleaning NaN values 

```python
# Assign default values for each data type
defaultInt = 0
defaultString = 'NaN'
defaultFloat = 0.0

# Create lists by data tpe
intFeatures = ['Age']
stringFeatures = ['Gender', 'Country', 'self_employed', 'family_history', 'treatment', 'work_interfere',
                 'no_employees', 'remote_work', 'tech_company', 'anonymity', 'leave', 'mental_health_consequence',
                 'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview',
                 'mental_vs_physical', 'obs_consequence', 'benefits', 'care_options', 'wellness_program',
                 'seek_help']
floatFeatures = []

# Clean the NaN's
for feature in health:
    if feature in intFeatures:
        health[feature] = health[feature].fillna(defaultInt)
    elif feature in stringFeatures:
        health[feature] = health[feature].fillna(defaultString)
    elif feature in floatFeatures:
        health[feature] = health[feature].fillna(defaultFloat)
    else:
        print('Error: Feature %s not recognized.' % feature)
health.head(5)   
```

```python
#clean 'Gender'
#Slower case all columm's elements
gender = health['Gender'].str.lower()
#print(gender)

#Select unique elements
gender = health['Gender'].unique()

#Made gender groups
male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "Cis Male", "cis male"]
trans_str = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means"]           
female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]

for (row, col) in health.iterrows():

    if str.lower(col.Gender) in male_str:
        health['Gender'].replace(to_replace=col.Gender, value='Male', inplace=True)

    if str.lower(col.Gender) in female_str:
        health['Gender'].replace(to_replace=col.Gender, value='Female', inplace=True)

    if str.lower(col.Gender) in trans_str:
        health['Gender'].replace(to_replace=col.Gender, value='Transgender', inplace=True)

#Get rid of bullshit
stk_list = ['A little about you', 'p']
health = health[~health['Gender'].isin(stk_list)]

print(health['Gender'].unique())
```

```python
#complete missing age with mean
health['Age'].fillna(health['Age'].median(), inplace = True)

# Fill with median() values < 18 and > 120
s = pd.Series(health['Age'])
s[s<18] = health['Age'].median()
health['Age'] = s
s = pd.Series(health['Age'])
s[s>120] = health['Age'].median()
health['Age'] = s

#Ranges of Age
health['age_range'] = pd.cut(health['Age'], [0,20,30,65,100], labels=["0-20", "21-30", "31-65", "66-100"], include_lowest=True)
```

```python
#There are only 0.014% of self employed so let's change NaN to NOT self_employed
#Replace "NaN" string from defaultString

health['self_employed'] = health['self_employed'].replace([defaultString], 'No')
print(health['self_employed'].unique())
```

```python
#There are only 0.20% of self work_interfere so let's change NaN to "Don't know
#Replace "NaN" string from defaultString

health['work_interfere'] = health['work_interfere'].replace([defaultString], 'Don\'t know' )
print(health['work_interfere'].unique())
```

# Step 3 : Encoding Data

```python
#Encoding data

labelDict = {}
for feature in health:
    le = preprocessing.LabelEncoder()
    le.fit(health[feature])
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    health[feature] = le.transform(health[feature])
    
    # Get labels
    labelKey = 'label_' + feature
    labelValue = [*le_name_mapping]
    labelDict[labelKey] =labelValue
    
for key, value in labelDict.items():     
    print(key, value)

health.head()
```

## Testing if there is any more missing data

```python
#missing data

total = health.isnull().sum().sort_values(ascending=False)
percent = (health.isnull().sum()/health.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
print(missing_data)
```
